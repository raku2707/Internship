## 1. Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
#display all the header tags from wikipedia.org and make data frame

header_tags=requests.get("https://en.wikipedia.org/wiki/Main_Page")
header_tags
```

```python
soup=BeautifulSoup(header_tags.content)
soup
```

```python
header=soup.find('span',class_='mw-headline')
header

header.text
```

## 9. Write a python program to scrape mentioned details from dineout.co.in and make data frame.
    i) Restaurant name
    ii) Cuisine
    iii) Location
    iv) Ratings
    v) Image URL

```python
#send the requestto the webpage server to get the source code of the page

page = requests.get("https://www.dineout.co.in/surat-restaurants?city_name=surat&limit=21&search_needle=restaurant&start=0&cityId=62&arr_locarea=Kamrej%2C%20East%20Surat&listing=1&tag%5B%5D=Top%20Budget%20Friendly&showAvailableTicket=0")
page
```

```python
#Page content
soup=BeautifulSoup(page.content)
soup
```

```python
#Srapping all restaurant name

restaurant_name = []

for i in soup.find_all('a', class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)
    
restaurant_name
```

```python
#Scrapping all cuisine 

cuisine = []

for i in soup.find_all('div', class_="restnt-info cursor"):
    cuisine.append(i.text)
    
cuisine
```

```python
#Srapping all locations name

location = []

for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

```python
#Srapping all restaurants rating

rating = []

for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rating.append(i.text)
    
rating
```

```python
#Srapping all image url

images = []

for i in soup.find_all('img', class_="no-img"):
    images.append(i.get('data-src'))
    
images
```

```python
#print length

print(len(restaurant_name),len(cuisine),len(location),len(rating),len(images))
```

```python
#Making dataframe

import pandas as pd
df = pd.DataFrame({'Restaurant_name':restaurant_name,'Cuisine':cuisine,'Location':location,'Rating':rating,'Images_url':images})
df
```
